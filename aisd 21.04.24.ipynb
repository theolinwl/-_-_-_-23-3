   1. Создать скрипт для анализа тестовых файлов - сбора статистики упоминания слов. Аргументом командной строки является имя файла, который нужно проанализировать. Второй аргумент - количество слов, которые нужно выводить на экран (если аргумент не указан, то выводим статистику по 100 словам, статистика выводится в порядке убывания частоты слов, по каждому слову на экран выводится строка в формате "<слово> <частота>"). Если второй аргумент не целое число, то его рассматриваем как имя файла, в который в формате CSV с заголовком сохраняем статистику слов (Первый столбец - слово, второй столбце - частота).

```python
def script(file_name,count=100):
    import csv
    import string
    dictt={}
    # открываем файл
    with open(file_name, 'r',encoding='utf-8') as file:
        file=file.read()
    # убираем все пунктуационные знаки
    for i in string.punctuation:
        file=file.replace(i,'')
        file=file.lower()
    file=file.replace('…','')
    file=file.replace('–','')
    file=file.split()
    # проходимся по файлу и добавляем слово в словарь, если оно встречается впервые, или увеличиваем частоту на 1
    for i in file:
            if i in dictt:
                dictt[i]+=1
            else:
                dictt[i]=1
    # проверяем целый ли второй аргумент
    if type(count)==int:
        # если целый, то сортируем и сохраняем в словарь нужное количество слов
        dictt=dict(sorted(dictt.items(),key=lambda item: item[1],reverse=True)[:count+1]) 
        for key,value in dictt.items():
            print(key,value)
    else:       
        # если не целое, то записываем словарь в новый файл
        dictt=dict(sorted(dictt.items(),key=lambda item: item[1],reverse=True)) 
        with open(str(count)+'.csv','w',newline='') as csvfile:
            writer=csv.writer(csvfile)
            for key, value in dictt.items():
                writer.writerow([key, value])

if __name__ == '__main__':
    import sys
    # проверяем, сколько аргументов введено
    if len(sys.argv) < 2:
        file_name = sys.argv[1]
    else:
        file_name = sys.argv[1]
        if type(sys.argv[2])==int:
            count = int(sys.argv[2])
        else:
            count=sys.argv[2]
script('shinel.txt',250)
```

```python
def total_word_frequency(*filenames):
    import csv
    import os
    #проверяем существует ли файл с помощью os
    for i in filenames:
        if not os.path.isfile(i):
            raise ValueError(f"Файл {i} не существует")
    dictt={}
    for i in args:
        # считываем каждый файл
        with open(i,'r') as file:
            #пропускаем одну строчку
            next(file)
            file=csv.reader(file)
            #проверяем, есть ли слово уже в словаре
            #если есть, то добавляем частоту к текущей
            #и если нет, добавляем слово и частоту
            for j in file:
                if j[0] in dictt:
                    dictt[j[0]]+=int(j[1])
                else:
                    dictt[j[0]]=int(j[1])
    file_name=input('Введите имя файла, в который сохраняется результат ')+'.csv'
    #сохранем результат в новый файл
    try:
        with open(file_name,'w',newline='') as result:
            writer=csv.writer(result)
            for key, value in dictt.items():
                writer.writerow([key,value])
    except:
        print('Ошибка при записи файла')
if __name__ == '__main__':
    import sys
    # проверяем, сколько аргументов введено
    if len(sys.argv) < 1:
        print('Не переданы имена файлов')
        sys.exit(1)
    else:
        filenames=sys.argv
```

```python
#программа, которая сохранена в модуле calculate.py
def calc(s):
    import math
    # Библиотеки для перевод чисел в строки:
    numb = {'ноль': '0', 'один': '1', 'два': '2', 'три': '3', 'четыре': '4', 'пять': '5', 'шесть': '6', 'семь': '7',
        'восемь': '8', 'девять': '9', 'десять': '10', 'двадцать': '20', 'тридцать': '30', 'сорок': '40',
        'пятьдесят': '50', 'шестьдесят': '60', 'семьдесят': '70', 'восемьдесят': '80', 'девяносто': '90',
        'одиннадцать': '11',
        'двенадцать': '12', 'тринадцать': '13', 'четырнадцать': '14', 'пятнадцать': '15', 'шестнадцать': '16',
        'семнадцать': '17', 'восемнадцать': '18', 'девятнадцать': '19'}
    des = {'двадцать': '20', 'тридцать': '30', 'сорок': '40', 'пятьдесят': '50', 'шестьдесят': '60', 'семьдесят': '70',
       'восемьдесят': '80', 'девяносто': '90'}
    s = s.split()
    k, l = '', ''
    i = 0
    if s[0] in numb or s[0] in des:
        while i < len(s):
            if i + 1 < len(s):
                if s[i] in numb:
                    if s[i] in des and s[i + 1] not in ('+', '-', '*'):
                        l = int(des[s[i]])
                    else:
                        if l == '':
                            k += numb[s[i]]
                        else:
                            k += str(l + int(numb[s[i]]))
                            l = ''
                if s[i] in '+-*':
                    k+=s[i]
            else:
                k += numb[s[i]]
            i += 1
    else:
        s=' '.join(s)
        k=s.replace(' ','')
    res = eval(k)
    return res

print(calc(input()))
```
